```python
import pandas as pd
import plotly as py
import plotly.graph_objs as go
```

```python
df1 = pd.read_csv('KL_value_90_lesser_regress.csv')
df2 = pd.read_csv('KL_value_90_greater_regress.csv')

df = pd.concat([df1, df2])
df.reset_index(drop=True, inplace=True)
del df['Unnamed: 0']
# print (df)

trace1 = go.Scatter(
    x = df['Metric'],
    y = df['Distance_Metric'],
    mode = 'markers',
    name = 'Scatter'
)

trace2 = go.Scatter(
    x = df['Metric'],
    y = df['Regress_Output'],
    mode = 'lines',
    name = 'Regressed Line'
)

data = [trace1, trace2]

py.offline.init_notebook_mode(connected=True)

xaxis=dict(title='x Axis')

py.offline.iplot({
    "data": data,
    "layout": go.Layout(title="KL Divergence Scatter (Rho : -0.849, 0.646)", xaxis=dict(title='Fraction Cases for which (Duration > 3 months)'), yaxis=dict(title='KL Divergence'))},
    image ='png',
    image_width = 1000,
    image_height = 600
)
```
